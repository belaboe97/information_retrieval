# Search engine based on user profiles

In order to create user profiles, different approaches can be considerd. Easier but might less effective approaches are so called wordbags. The user intrests are encoded in bags of words which are related to each other. (Topic Modelling)

While this approach lacks an underlying understanding of the words relation, its seems to be quite sufficient for the rare modelling of user profile: 

Profile 1: Soccer
Profile 2: Movies
Profile 3: Music 
Profile 4: Politics

Other approaches could include so called Onthologies. Onthologies provide a detailed view how different topics and subtopics relate to each other. 

Checkout:https://www.researchgate.net/figure/The-process-of-generating-user-interest-profiles_fig1_308765723 

To create user profiles based on wordclouds, we can use the NLTK library. 

See: https://www.holisticseo.digital/python-seo/nltk/wordnet

# User profile
<br/>
As already explained, we use nltk library to create the user profile first (wordcloud). Therefore we are intressted in similiar words and words related to the topic. 
<br/> 
In order to generalise the approach, we define functions

1. Synonyms: Words that have the same meaning as the given Word

```python
#code obtained and modified from: https://www.holisticseo.digital/python-seo/nltk/wordnet , Chapter 
from nltk import corpus

def find_synonyms(inp : str) -> list: 
    synonyms = list()
    for syn in corpus.wordnet.synsets("soccer"):
        for i in syn.lemmas():
            synonyms.append(i.name())
    return list(set(synonyms))




```
